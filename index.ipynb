ID: This could refer to a unique identifier for each patient or sample in the dataset.

Diagnosis: This likely indicates the diagnosis of each sample, such as whether it is classified as malignant or benign.

radius_mean: The average radius of the detected tumor cells.

texture_mean: The average texture (variation in grayscale intensity) of the tumor cells.

perimeter_mean: The average perimeter (the length of the boundary) of the tumor cells.

area_mean: The average area (size) of the tumor cells.

smoothness_mean: The average smoothness of the boundary of the tumor cells.

compactness_mean: The average compactness of the tumor cells, which can be a measure of how closely packed the cells are.

concavity_mean: The average severity of concave portions of the contour of the tumor cells.

concave_points_mean: The average number of concave portions of the contour of the tumor cells.

symmetry_mean: The average symmetry of the tumor cells.

fractal_dimension_mean: The average fractal dimension, which is a measure of how complex the boundary of the tumor cells is.

radius_se: The standard error of the radius of the tumor cells.

texture_se: The standard error of the texture of the tumor cells.

perimeter_se: The standard error of the perimeter of the tumor cells.

area_se: The standard error of the area of the tumor cells.

smoothness_se: The standard error of the smoothness of the boundary of the tumor cells.

compactness_se: The standard error of the compactness of the tumor cells.

concavity_se: The standard error of the severity of concave portions of the contour of the tumor cells.

concave_points_se: The standard error of the number of concave portions of the contour of the tumor cells.

symmetry_se: The standard error of the symmetry of the tumor cells.

fractal_dimension_se: The standard error of the fractal dimension of the tumor cells.

radius_worst: The "worst" or largest radius among the tumor cells.

texture_worst: The "worst" or most intense texture among the tumor cells.

perimeter_worst: The "worst" or largest perimeter among the tumor cells.

area_worst: The "worst" or largest area among the tumor cells.

smoothness_worst: The "worst" or least smooth boundary among the tumor cells.

compactness_worst: The "worst" or highest compactness among the tumor cells.

concavity_worst: The "worst" or highest severity of concave portions of the contour among the tumor cells.

concave_points_worst: The "worst" or highest number of concave portions of the contour among the tumor cells.

symmetry_worst: The "worst" or least symmetric tumor cells.

fractal_dimension_worst: The "worst" or highest fractal dimension among the tumor cells.

```python
def normalize_data_frame(data_frame):
    """
    Purpose
    ----------
    Function created to normalize data set.
    Intializes an empty data frame which will normalize all columns that
    have at > 10 unique values (chosen arbitrarily since target columns
    will have classes < 10) and append the non-float types.
    Application can vary significantly for different data set, use with caution
    or modify accordingly.

    Parameters
    ----------
    * data_frame: 	Name of pandas.dataframe

    Returns
    ----------
    * data_frame_norm:	Normalized dataframe values ranging (0, 1)
    """
    data_frame_norm = pd.DataFrame()
    for col in data_frame:
        if ((len(np.unique(data_frame[col])) > 10) & (data_frame[col].dtype != 'object')):
            data_frame_norm[col]=((data_frame[col] - data_frame[col].min()) /
                                  (data_frame[col].max() - data_frame[col].min()))
        else:
            data_frame_norm[col] = data_frame[col]
    return data_frame_norm
```
